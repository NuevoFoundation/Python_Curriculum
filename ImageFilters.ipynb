## Kitty pics

```python
# With the Image.open() function we can view our kitty image.
Image.open('images/cat.jpg')
```

### Black and white kitty
Using the PIL library, let's write a function that turns our orange kitty to black and white.

The black and white mode is indicated by an 'L'.

Use the .convert() function to change the image and indicate that you want it in black and white by passing the 'L' to the function like this: 

color_image.convert('L')

```python
orange_kitty = Image.open('images/cat.jpg')

black_and_white_kitty = orange_kitty.convert('L')
black_and_white_kitty.save('bw_cat.jpg')

Image.open('bw_cat.jpg')
```

### Blurry kitty

Let's make our kitty blurry using the BLUR filter.

The steps to apply the filter are:
1. Open the image
2. Apply the filter to the image using the filter('ImageFilter.BLUR') function
3. Save the image

```python
from PIL import ImageFilter
```

```python
# 1. Open the orange kitty image
orange_kitty = Image.open('images/cat.jpg')

# 2. Apply the BLUR filter
blurry_kitty = orange_kitty.filter(ImageFilter.BLUR)

# Save our blurry kitty image
blurry_kitty.save('blurry-kitty.jpg')
```

Let's see what our blurry kitty looks like:

```python
Image.open('blurry-kitty.jpg')
```
